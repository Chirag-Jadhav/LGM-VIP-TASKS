Image to pencil sketch

```python
import cv2
from IPython.display import Image
from google.colab.patches import cv2_imshow
```

Getting Original Image

```python
image = cv2.imread("goldenretriever.jpg")
cv2_imshow(image)
```

Converting Image to Grey Scale

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

Invert the new gray scale

```python
inverted_image = 255 - gray_image
cv2_imshow(inverted_image)
```

Blur the image by using the Gaussian Function in OpenCV

```python
blur_image = cv2.GaussianBlur(inverted_image,(21,21),0)
cv2_imshow(blur_image)
```

Invert the Blur Image

```python
inverted_blur_image = 255 - blur_image
cv2_imshow(inverted_blur_image)
```

pencil Sketch

```python
pencil_sketch = cv2.divide(gray_image, inverted_blur_image, scale=256.0)
cv2_imshow(pencil_sketch)
```

Display Both Original and Pencil Sketch

```python
cv2_imshow(image)
cv2_imshow(pencil_sketch)
```
